```python
import time
import matplotlib.pyplot as plt

def isort(arr,n):
  for i in range(1,n):
    temp = arr[i]
    j = i -1
    while j >=0 and arr[j] > temp:
      arr[j + 1] = arr[j]
      j -= 1
    arr[j+ 1] = temp

def ssort(arr,n):
  for i in range(0,n-1):
    min = i
    for j in range(1, n):
      if arr[j] < arr[min]:
        min = j

    if min != i:
      arr[i], arr[min] = arr[min],arr[i]


def generate_k_nearly_sorted_array(n, k):
    arr = list(range(n))
    for i in range(n):
        j = min(i + k, n - 1)
        arr[i], arr[j] = arr[j], arr[i]
    return arr

sizes = [1000, 5000, 10000, 20000, 50000]
k = 10
insertion_times = []
bubble_times = []

for size in sizes:
    arr = generate_k_nearly_sorted_array(size, k)

    # Measure time for Insertion Sort
    start_time = time.time()
    isort(arr.copy() , len(arr))
    insertion_times.append(time.time() - start_time)

    # Measure time for Bubble Sort
    start_time = time.time()
    ssort(arr.copy() , len(arr))
    bubble_times.append(time.time() - start_time)

# Plotting the results
plt.plot(sizes, insertion_times, label="Insertion Sort")
plt.plot(sizes, bubble_times, label="Selection Sort")
plt.xlabel('Array Size')
plt.ylabel('Time (seconds)')
plt.title('Time Complexity of Sorting Algorithms on k-Nearly Sorted Array')
plt.legend()
plt.show()

```

```python
import time
import matplotlib.pyplot as plt

def binary_search(arr, key, start, end):
    """
    Performs a binary search to find the index where 'key' should be inserted.
    """
    while start <= end:
        mid = (start + end) // 2
        if arr[mid] < key:
            start = mid + 1
        else:
            end = mid - 1
    return start

def improved_insertion_sort(arr):
    """
    Sorts the array using the improved insertion sort algorithm that uses binary search
    to find the correct position for each element.
    """
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        pos = binary_search(arr, key, 0, i - 1)  # Find the position using binary search
        # Shift elements to the right to make space for the key
        arr = arr[:pos] + [key] + arr[pos:i] + arr[i+1:]
    return arr


def generate_arr(s):
  arr = list(range(s))
  for i in range(s):
    j = min(i + 10 , s - 1)
    arr[i] ,arr[j] = arr[j] ,arr[i]
  return arr

def isort(arr,n):
  for i in range(1,n):
    temp = arr[i]
    j = i- 1
    while j >= 0 and arr[j] > temp:
      arr[j+ 1] = arr[j]
      j -= 1
    arr[j + 1] = temp
  return arr

sizes = [200,300,500,600]

isort_times = []
imsort = []

for size in sizes:
  arr = generate_arr(size)
  start_time = time.time()
  isort(arr,len(arr))
  isort_times.append(time.time() - start_time)

  start_time = time.time()
  improved_insertion_sort(arr)
  imsort.append(time.time() - start_time)


plt.plot(sizes,isort_times)
plt.plot(sizes,imsort)
plt.legend()
plt.show()
# Example usage
# arr = [7, 6, 10, 5, 9, 2, 1, 15, 7]
arr = [7, 6, 10, 5, 9, 2, 1, 15, 7]
sorted_arr = improved_insertion_sort(arr)
print("Sorted Array:", sorted_arr)
```

```python

def MinHeapify(arr, n, i):
    """
    Heapify a subtree rooted at index i.

    :param arr: The array to be heapified.
    :param n: The size of the heap.
    :param i: The index of the root node.
    """
    smallest = i
    lchild = 2 * i + 1
    rchild = 2 * i + 2

    # See if left child of root exists and is smaller than root
    if lchild < n and arr[lchild] < arr[smallest]:
        smallest = lchild

    # See if right child of root exists and is smaller than the current smallest
    if rchild < n and arr[rchild] < arr[smallest]:
        smallest = rchild

    # Change root, if needed
    if smallest != i:
        arr[i], arr[smallest] = arr[smallest], arr[i]
        MinHeapify(arr, n, smallest)

def MinHeap(arr, n):
    """
    Build a min heap.

    :param arr: The array to be heapified.
    :param n: The size of the array.
    """
    # Build a min heap
    for i in range(n // 2 - 1, -1, -1):
        MinHeapify(arr, n, i)

def MinHeapSort(arr, n):
    """
    Sort an array using the Min Heap Sort algorithm.

    :param arr: The array to be sorted.
    :param n: The size of the array.
    """
    # Build a min heap
    MinHeap(arr, n)

    # Extract elements from the heap one by one
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        MinHeapify(arr, i, 0)

# Example usage
arr = [12, 11, 13, 5, 6, 7]
n = len(arr)
MinHeapSort(arr, n)
print("Sorted array is:", arr)
```

```python
def MinHeapify(arr, n, i):

    smallest = i
    lchild = 2 * i + 1
    rchild = 2 * i + 2

    # See if left child of root exists and is smaller than root
    if lchild < n and arr[lchild][0] < arr[smallest][0]:
        smallest = lchild

    # See if right child of root exists and is smaller than the current smallest
    if rchild < n and arr[rchild][0] < arr[smallest][0]:
        smallest = rchild

    # Change root, if needed
    if smallest != i:
        arr[i], arr[smallest] = arr[smallest], arr[i]
        MinHeapify(arr, n, smallest)

def MergeSortedArrays(arrays):
    """
    Merge m sorted arrays using a Min Heap.

    :param arrays: A list of m sorted arrays.
    :return: A single sorted array.
    """
    m = len(arrays)
    n = sum(len(array) for array in arrays)
    result = [0] * n
    index = 0

    # Create a Min Heap to store the first element of each array
    min_heap = [(array[0], i, 0) for i, array in enumerate(arrays) if array]
    # print(min_heap)
    MinHeapify(min_heap, len(min_heap), 0)

    while min_heap:
        # Extract the smallest element from the Min Heap
        val, array_idx, elem_idx = min_heap[0]
        result[index] = val
        index += 1

        # Replace the extracted element with the next element from the same array
        if elem_idx + 1 < len(arrays[array_idx]):
            min_heap[0] = (arrays[array_idx][elem_idx + 1], array_idx, elem_idx + 1)
            MinHeapify(min_heap, len(min_heap), 0)
        else:
            # If the array is exhausted, remove it from the Min Heap
            min_heap.pop(0)
            MinHeapify(min_heap, len(min_heap), 0)

    return result

# Example usage
arrays = [[1, 3, 5], [2, 4, 6], [0, 7, 8, 9]]
result = MergeSortedArrays(arrays)
print("Merged sorted array is:", result)
```

```python
def find_k_largest(arr, k):
    """
    Find the K largest elements in an array of size N, where K << N.

    :param arr: The input array.
    :param k: The number of largest elements to find.
    :return: A list of the K largest elements.
    """
    # Create a min heap to store the K largest elements
    min_heap = [float('-inf')] * k

    # Build the min heap
    for i in range(k):
        MinHeapify(min_heap, k, i)

    # Iterate through the array
    for num in arr:
        # If the current number is larger than the root of the min heap
        if num > min_heap[0]:
            # Replace the root with the current number
            min_heap[0] = num
            # Heapify the min heap
            MinHeapify(min_heap, k, 0)

    # Return the K largest elements
    return min_heap

def MinHeapify(arr, n, i):
    """
    Heapify a subtree rooted at index i.

    :param arr: The array to be heapified.
    :param n: The size of the heap.
    :param i: The index of the root node.
    """
    smallest = i
    lchild = 2 * i + 1
    rchild = 2 * i + 2

    # See if left child of root exists and is smaller than root
    if lchild < n and arr[lchild] < arr[smallest]:
        smallest = lchild

    # See if right child of root exists and is smaller than the current smallest
    if rchild < n and arr[rchild] < arr[smallest]:
        smallest = rchild

    # Change root, if needed
    if smallest != i:
        arr[i], arr[smallest] = arr[smallest], arr[i]
        MinHeapify(arr, n, smallest)

# Example usage
arr = [5, 2, 8, 3, 1, 6, 4]
k = 3
result = find_k_largest(arr, k)
print("The", k, "largest elements are:", result)
```

```python
def max_activities(activities):
    """
    Find the maximum number of activities that can be performed by a single person.

    :param activities: A list of tuples, where each tuple contains the start and end times of an activity.
    :return: The maximum number of activities that can be performed.
    """
    # Sort the activities by their end times
    activities.sort(key=lambda x: x[1])

    # Initialize the maximum number of activities and the end time of the last activity
    max_activities = 1
    last_end_time = activities[0][1]

    # Iterate through the activities
    for start, end in activities[1:]:
        # If the current activity starts after the last activity ends, increment the count
        if start >= last_end_time:
            max_activities += 1
            last_end_time = end

    return max_activities

# Example usage
activities = [(1, 4), (3, 5), (0, 6), (5, 7), (3, 8), (5, 9), (6, 10), (8, 11), (8, 12), (2, 13), (12, 14)]
result = max_activities(activities)
print("The maximum number of activities is:", result)
```

```python
import math

def introsort(arr):
    def introsort_util(arr, lb, ub, depth_limit):
        size = ub - lb + 1
        if size < 16:
            isort(arr, lb, ub)
        elif depth_limit == 0:
            HeapSort(arr, lb, ub)
        else:
            loc = Partition(arr, lb, ub)
            introsort_util(arr, lb, loc - 1, depth_limit - 1)
            introsort_util(arr, loc + 1, ub, depth_limit - 1)

    def isort(arr, lb, ub):
        for i in range(lb + 1, ub + 1):
            temp = arr[i]
            j = i - 1
            while j >= lb and arr[j] > temp:
                arr[j + 1] = arr[j]
                j -= 1
            arr[j + 1] = temp

    def Partition(arr, lb, ub):
        pivot = arr[lb]
        start = lb
        end = ub

        while start < end:
            while start < len(arr) and arr[start] <= pivot:
                start += 1
            while arr[end] > pivot:
                end -= 1
            if start < end:
                arr[start], arr[end] = arr[end], arr[start]

        arr[lb], arr[end] = arr[end], arr[lb]
        return end

    def HeapSort(arr, lb, ub):
        n = ub - lb + 1
        # Build a max heap
        for i in range(n // 2 - 1, -1, -1):
            Heapify(arr, lb, ub, i)

        # One by one extract elements
        for i in range(n - 1, 0, -1):
            arr[lb], arr[lb + i] = arr[lb + i], arr[lb]  # Swap
            Heapify(arr, lb, lb + i - 1, 0)  # Heapify the root element to get the largest at root again

    def Heapify(arr, lb, ub, i):
        largest = i  # Initialize largest as root
        lchild = 2 * i + 1  # Left child index
        rchild = 2 * i + 2  # Right child index

        # See if left child of root exists and is greater than root
        if lchild <= ub - lb and arr[lb + lchild] > arr[lb + largest]:
            largest = lchild

        # See if right child of root exists and is greater than the current largest
        if rchild <= ub - lb and arr[lb + rchild] > arr[lb + largest]:
            largest = rchild

        # Change root, if needed
        if largest != i:
            arr[lb + i], arr[lb + largest] = arr[lb + largest], arr[lb + i]  # Swap
            Heapify(arr, lb, ub, largest)  # Recursively heapify the affected sub-tree

    # Initial call to introsort_util
    depth_limit = 2 * int(math.log2(len(arr)))
    introsort_util(arr, 0, len(arr) - 1, depth_limit)

# Example usage:
arr = [10, 7, 8, 9, 1, 5]
introsort(arr)
print("Sorted array:", arr)
```

```python
def find_pairs(arr, target_sum):
    """
    Find all pairs in an unsorted int array that sum up to a given target sum.

    :param arr: The input array.
    :param target_sum: The target sum.
    :return: A list of tuples, where each tuple contains a pair of numbers that sum up to the target sum.
    """
    # Create a hashmap to store the numbers we've seen so far and their indices
    num_indices = {}

    # Create a list to store the result
    result = []

    # Iterate through the array
    for i, num in enumerate(arr):
        # Calculate the complement of the current number
        complement = target_sum - num

        # Check if the complement is in the hashmap
        if complement in num_indices:
            # Add the pair to the result
            result.append((complement, num))

        # Add the current number to the hashmap
        num_indices[num] = i

    return result

# Example usage
arr = [8, 7, 2, 5, 3, 1]
target_sum = 10
result = find_pairs(arr, target_sum)
print("The pairs that sum up to", target_sum, "are:", result)
```
